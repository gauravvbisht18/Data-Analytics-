```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
        
df=pd.read_csv('/kaggle/input/heart-disease/heart.csv')
df.head()
```

# Distribution PLot for Numeric Value

```python
sns.displot(df['Age'], kde = True, color = 'red')
```

```python
sns.displot(x = df['RestingBP'], kde = True, color = 'red')
```

```python
sns.displot(df['Cholesterol'], kde = True, color = 'red')
```

```python
sns.displot(df['MaxHR'], kde = True, color = 'red')
```

# Pie Charts for Categorical Value

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

# Correlation JointPlot

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Cholesterol' , y = 'MaxHR', data = df, kind = 'reg')
```

```python
sns.jointplot(x = 'HeartDisease' , y = 'MaxHR', data = df, kind = 'reg')
```
